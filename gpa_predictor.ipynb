# Imports


```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import pickle
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
```

# Preprocessing

```python
# function for preprocessing
def preprocess():
    df=pd.read_csv("The_Grades_Dataset.csv") #read dataset into Dataframe
    target=df.CGPA  #separating CGPA column as target
    df.drop(["Seat No.","CGPA"],axis="columns",inplace=True) #dropping CGPA from input dataframe
    #encoding
    scale_mapper = {"A+":1, "A":2, "A-":3, "B+":4, "B":5, "B-":6, "C+":7, "C":8, "C-":9, "D+":10,"D":11,"F":12,"I":13,"W":14,"WU":15} 
    df=df.replace(scale_mapper) 
    df.fillna(8,inplace=True) #replace missing values with 8
    return df , target
```

# Linear Model

```python
def Linear_model(df,target,num):
    df=df
    target=target
    num=num
    
    model=LinearRegression()  #calling LinearRegression 
    X = df.iloc[:, :].values
    X_train,X_test,Y_train,Y_test = train_test_split(X,target,test_size=0.2,random_state=0)
    model.fit(X_train,Y_train)#fitting inputs and target into the model
    #metrics
#     r_sq = model.score(X, target)
    train_score = model.score(X_train, Y_train)
    test_score = model.score(X_test, Y_test)
#     print(f"coefficient of determination of entire data: {r_sq}")
    print(f"coefficient of determination of training data using Linear Regression: {train_score}")
    print(f"coefficient of determination of test data using Linear Regression: {test_score}")
    print(f"intercept of the model: {model.intercept_}")
    print(f"slope of the model: {model.coef_}")
    pickle.dump(model,open( 'model'+num+'.pkl', 'wb'))  #dumping model in pickle file to save in harddisk from memory
    return model
```

# SVR Model

```python
def svr_model(df,target,num):
    df=df
    X = df.iloc[:, :].values
    target=target
    num=num
    X_train,X_test,Y_train,Y_test = train_test_split(X,target,test_size=0.2,random_state=0)
    
    svr_rbf=SVR(C=1.0, epsilon=0.1, kernel='rbf')  #calling LinearRegression 
    svr_rbf.fit(X_train, Y_train)  #fitting inputs and target into the model
#     r_sq = svr_rbf.score(X, target)
    train_score = svr_rbf.score(X_train, Y_train)
    test_score = svr_rbf.score(X_test, Y_test)
#     print(f"coefficient of determination of entire data using SVR: {r_sq}")
    print(f"coefficient of determination of training data using SVR: {train_score}")
    print(f"coefficient of determination of test data using SVR: {test_score}")
   
    pickle.dump(svr_rbf,open( 'model_'+num+'.pkl', 'wb'))  #dumping model in pickle file to save in harddisk from memory  
    return svr_rbf
```
